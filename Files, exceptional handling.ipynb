```python
# Q 4 .

def copy_file_content(source_file_path, destination_file_path):
    try:
        # Open the source file in read mode
        with open(source_file_path, 'r') as source_file:
            content = source_file.read()

        # Open the destination file in write mode
        with open(destination_file_path, 'w') as destination_file:
            destination_file.write(content)

        print("Content successfully copied from source to destination file.")
    except FileNotFoundError:
        print(f"Error: The file '{source_file_path}' does not exist.")
    except IOError as e:
        print(f"Error: An I/O error occurred. Details: {e}")

# Example usage
source_file_path = 'source.txt'
destination_file_path = 'destination.txt'
copy_file_content(source_file_path, destination_file_path)

```

```python
# Q 9
# ANS.

# Function to read file content line by line and store it in a list
def read_file_to_list(file_path):
    content_list = []
    try:
        with open(file_path, 'r') as file:
            for line in file:
                content_list.append(line.strip())  # Strip newline characters and add to the list
        return content_list
    except FileNotFoundError:
        print("Error: The file does not exist.")
        return []
    except IOError as e:
        print(f"Error: An I/O error occurred. Details: {e}")
        return []

# Example usage
file_path = 'example.txt'
lines = read_file_to_list(file_path)
print(lines)
```

```python
# Q 10

# Function to append data to an existing file
def append_to_file(file_path, data):
    try:
        # Open the file in append mode
        with open(file_path, 'a') as file:
            # Write the data to the file
            file.write(data + '\n')  # Adding a newline character for each entry
        print("Data appended successfully.")
    except IOError as e:
        print(f"Error: An I/O error occurred. Details: {e}")

# Example usage
file_path = 'example.txt'
data_to_append = 'This is the new data to be appended.'
append_to_file(file_path, data_to_append)
```

```python
#Q 11

# Function to get a value from a dictionary by key
def get_value_from_dict(dictionary, key):
    try:
        # Attempt to access the dictionary key
        value = dictionary[key]
        return value
    except KeyError:
        # Handle the case where the key doesn't exist
        print(f"Error: The key '{key}' does not exist in the dictionary.")
        return None

# Example usage
my_dict = {'name': 'Ram', 'age': 30, 'city': 'delhi'}
key_to_access = 'country'

value = get_value_from_dict(my_dict, key_to_access)
if value is not None:
    print(f"The value for the key '{key_to_access}' is {value}")
```

```python
# Q 12

# Function to demonstrate multiple exception handling
def exception_demo(a, b, index):
    try:
        # Division operation
        result = a / b
        print(f"Result of division: {result}")

        # Accessing a list element
        my_list = [1, 2, 3]
        element = my_list[index]
        print(f"Element at index {index}: {element}")

    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")

    except IndexError:
        print(f"Error: Index {index} is out of range.")

    except TypeError:
        print("Error: Invalid input type.")

# Example usage
exception_demo(10, 0, 1)
exception_demo(10, 2, 5)
exception_demo('10', 2, 1)


```

```python
# Q 14

import logging

# Configure the logger
logging.basicConfig(level=logging.DEBUG,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# Log an informational message
logging.info('Program started')

def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        # Log an error message
        logging.error('Division by zero!')
        return None
    else:
        return result

# Perform a division that will be logged as info
result1 = divide(10, 2)
logging.info(f'Result of 10 / 2: {result1}')


# Perform a division that will be logged as error
result2 = divide(10, 0)
logging.info(f'Result of 10 / 0: {result2}')


# Log another informational message
logging.info('Program finished')
```

```python
# Q 17

def write_numbers_to_file(file_path, numbers):
    try:
        with open(file_path, 'w') as file:
            for number in numbers:
                file.write(f"{number}\n")
        print("Numbers successfully written to the file.")
    except IOError as e:
        print(f"Error: An I/O error occurred. Details: {e}")

# Example usage
file_path = 'numbers.txt'
numbers = list(range(1, 101))  # Create a list of numbers from 1 to 100
write_numbers_to_file(file_path, numbers)

```

```python
# Q 18

import logging
from logging.handlers import RotatingFileHandler

# Configure the logging system
log_file = 'app_log.txt'
max_log_size = 1 * 1024 * 1024  # 1MB
backup_count = 3  # Number of backup files to keep

# Create a rotating file handler
handler = RotatingFileHandler(log_file, maxBytes=max_log_size, backupCount=backup_count)

# Set the logging format
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Get the root logger and configure it
logger = logging.getLogger()
logger.setLevel(logging.DEBUG)  # Set the logging level to DEBUG to capture all levels of messages
logger.addHandler(handler)

# Example log messages
logger.debug('This is a debug message')
logger.info('This is an informational message')
logger.warning('This is a warning message')
logger.error('This is an error message')
logger.critical('This is a critical message')

print("Logging setup with rotation is complete. Check the log file for messages.")
```

```python
# Q 19

def handle_exceptions():
    # Example list and dictionary
    my_list = [10, 20, 30]
    my_dict = {'a': 1, 'b': 2}

    try:
        # Attempt to access an index that may cause an IndexError
        print(f"List element: {my_list[3]}")  # This will cause an IndexError

        # Attempt to access a dictionary key that may cause a KeyError
        print(f"Dictionary value: {my_dict['c']}")  # This will cause a KeyError

    except IndexError as e:
        print(f"IndexError: {e}")

    except KeyError as e:
        print(f"KeyError: {e}")

    # Additional code to run regardless of exceptions
    print("Exception handling completed.")

# Example usage
handle_exceptions()
```

```python
# Q21

def count_word_occurrences(file_path, target_word):
    try:
        with open(file_path, 'r') as file:
            content = file.read()

        # Convert the content to lowercase and split into words
        words = content.lower().split()

        # Count the occurrences of the target word
        word_count = words.count(target_word.lower())

        print(f"The word '{target_word}' occurs {word_count} times in the file.")
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' does not exist.")
    except IOError as e:
        print(f"Error: An I/O error occurred. Details: {e}")

# Example usage
file_path = 'example.txt'
target_word = 'specific'
count_word_occurrences(file_path, target_word)
```

```python
# Q22

import os

def is_file_empty(file_path):
    try:
        if os.path.getsize(file_path) == 0:
            print("The file is empty.")
        else:
            with open(file_path, 'r') as file:
                content = file.read()
                print("File content:")
                print(content)
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' does not exist.")
    except IOError as e:
        print(f"Error: An I/O error occurred. Details: {e}")

# Example usage
file_path = 'example.txt'
is_file_empty(file_path)
```
